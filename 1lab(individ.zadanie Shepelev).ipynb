```python
import pandas as pd

data2 = pd.read_csv('wine.data').values
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

class_label = []          # Class
alcohol = []              # Alcohol
malic_acid = []          # Malic acid
ash = []                  # Ash
alcalinity_of_ash = []    # Alcalinity of ash
magnesium = []             # Magnesium
total_phenols = []        # Total phenols
flavanoids = []           # Flavanoids
nonflavanoid_phenols = [] # Nonflavanoid phenols
proanthocyanins = []      # Proanthocyanins
color_intensity = []      # Color intensity
hue = []                  # Hue
od280_od315 = []          # OD280/OD315 of diluted wines
proline = []              # Proline

for dot in data2:
    class_label.append(dot[0])
    alcohol.append(dot[1])
    malic_acid.append(dot[2])
    ash.append(dot[3])
    alcalinity_of_ash.append(dot[4])
    magnesium.append(dot[5])
    total_phenols.append(dot[6])
    flavanoids.append(dot[7])
    nonflavanoid_phenols.append(dot[8])
    proanthocyanins.append(dot[9])
    color_intensity.append(dot[10])
    hue.append(dot[11])
    od280_od315.append(dot[12])
    proline.append(dot[13])

plt.figure(figsize=(15, 20))

plt.subplot(5, 2, 1)
plt.scatter(alcohol, malic_acid, c=class_label, cmap='viridis', label='Alcohol vs. Malic Acid')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 2)
plt.scatter(alcohol, total_phenols, c=class_label, cmap='viridis', label='Alcohol vs. Total Phenols')
plt.xlabel('Alcohol')
plt.ylabel('Total Phenols')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 3)
plt.scatter(alcohol, color_intensity, c=class_label, cmap='viridis', label='Alcohol vs. Color Intensity')
plt.xlabel('Alcohol')
plt.ylabel('Color Intensity')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 4)
plt.scatter(alcohol, flavanoids, c=class_label, cmap='viridis', label='Alcohol vs. Flavanoids')
plt.xlabel('Alcohol')
plt.ylabel('Flavanoids')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 5)
plt.scatter(alcohol, ash, c=class_label, cmap='viridis', label='Alcohol vs. Ash')
plt.xlabel('Alcohol')
plt.ylabel('Ash')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 6)
plt.scatter(alcohol, alcalinity_of_ash, c=class_label, cmap='viridis', label='Alcohol vs. Alcalinity of Ash')
plt.xlabel('Alcohol')
plt.ylabel('Alcalinity of Ash')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 7)
plt.scatter(alcohol, magnesium, c=class_label, cmap='viridis', label='Alcohol vs. Magnesium')
plt.xlabel('Alcohol')
plt.ylabel('Magnesium')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 8)
plt.scatter(alcohol, nonflavanoid_phenols, c=class_label, cmap='viridis', label='Alcohol vs. Nonflavanoid Phenols')
plt.xlabel('Alcohol')
plt.ylabel('Nonflavanoid Phenols')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 9)
plt.scatter(alcohol, proanthocyanins, c=class_label, cmap='viridis', label='Alcohol vs. Proanthocyanins')
plt.xlabel('Alcohol')
plt.ylabel('Proanthocyanins')
plt.colorbar(label='Class')
plt.legend()

plt.subplot(5, 2, 10)
plt.scatter(alcohol, od280_od315, c=class_label, cmap='viridis', label='Alcohol vs. OD280/OD315')
plt.xlabel('Alcohol')
plt.ylabel('OD280/OD315')
plt.colorbar(label='Class')
plt.legend()

plt.tight_layout()
plt.show()
```
